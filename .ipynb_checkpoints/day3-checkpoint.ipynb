```python
def replace_gender(cell):
    if cell == "Female":
        return "Women"
    elif cell == "Male":
        return "Men"

df3 = pd.read_excel('file_example.xls', sheet_name="Data", converters={"Gender": replace_gender})
df3
```

```python
df4 = pd.read_excel("file_example.xls", header=6)
df4
```

# Handling Missing Data

```python
df = df5.fillna("No Value")
df.head(10)
```

```python
df3 = df5.fillna({
    'Last Name': "No Last name",
    'Company': "No Company",
    'City': "No City",
    'Country': "No Country"
})
df3.head(10)
```

```python
df4 = df5.fillna(method='ffill')
df4.head(10)
```

```python
df4 = df5.fillna(method='bfill')
df4.head(10)
```

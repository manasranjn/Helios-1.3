```python
import pandas as pd
```

```python
temp = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru', 'Kolkata'],
    "temp": [12,13,19, 30]})
temp
```

```python
wind = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Delhi'],
    "wind": [5,6,23]})
wind
```

```python
df = pd.merge(temp, wind, on='City')
df
```

```python
df = pd.merge(temp, wind, on='City', how="left")
df
```

```python
df = pd.merge(temp, wind, on='City', how="right")
df
```

```python
df = pd.merge(temp, wind, on='City', how="outer")
df
```

```python
df = pd.merge(temp, wind, on='City', how="outer", indicator=True)
df
```

```python
df = pd.merge(temp, wind, on='City', how="left", indicator=True)
df
```

```python
df = pd.merge(temp, wind, on='City', how="left", suffixes=(False, False))
df
```

```python
data = {
    'Employee': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Department': ['HR', 'HR', 'HR', 'IT', 'IT', 'IT'],
    'Year': [2022, 2022, 2022, 2023, 2023, 2023],
    'Salary': [40000, 42000, 41000, 45000, 47000, 46000],
    'Bonus': [2000, 2100, 1900, 2500, 2700, 2600]
}

df = pd.DataFrame(data)
df
```

```python
pivot_df = df.pivot(index='Employee', columns='Year', values='Salary')
pivot_df
```

```python
pivot_table_df = df.pivot_table(
    index='Employee',
    columns='Year',
    values='Salary',
)
pivot_table_df
```

```python
pivot_table_df = df.pivot_table(
    index='Employee',
    columns='Year',
    values='Salary',
    aggfunc='sum',     
    fill_value=0,      
    margins=True,    
    margins_name='Total Salary'
)
pivot_table_df
```

```python
data = {
    'Employee': ['A', 'B', 'C', 'A', 'B', 'C'],
    'Department': ['HR', 'HR', 'HR', 'IT', 'IT', 'IT'],
    'Year': [2022, 2022, 2022, 2023, 2023, 2023],
    'Salary': [40000, 42000, 41000, 45000, 47000, 46000],
    'Bonus': [2000, 2100, 1900, 2500, 2700, 2600]
}

df = pd.DataFrame(data)
df
```

```python
melt_df = pd.melt(
    df,
    id_vars=['Employee', 'Department'],   
    value_vars=['Salary', 'Bonus', 'Year'],       
    var_name='Title',                  
    value_name='Amount'                   
)
melt_df
```

# From Dictionary

```python
person1 = {
    "name": ["Smith", "Allen", "John"],
    "age": [20,32,24],
    "job": ["SDE", "HR", "Sales"]
}
df1 = pd.DataFrame(person1)
df1
```

```python
df2 = pd.DataFrame({
    "name": ["Smith", "Allen", "John"],
    "age": [20,32,24],
    "job": ["SDE", "HR", "Sales"]
})
df2
```

# From List

```python
df3 = pd.DataFrame([[10,20,30,40], [1,2,3,4]])
df3
```

```python
df4 = pd.DataFrame([[10,20,30,40], [1,2,3,4]], columns=['A','B','C','D'])
df4
```

```python
arr1 = np.array([[10,20,30,40], [1,2,3,4]])
arr1
```
